### Q make func provide only even num in list 

```python
def filter_even(lst):
    ans = []
    # automatic iteration on list
    for number in lst:
        if number % 2 == 0:
            ans.append(number)
    return ans

lst = list(map(int,input("Enter ele of list : ").split()))
print(filter_even(lst))
```

```python
import math

def emi(principal,rate,duration,down_payment = 0):
    """Calculates the equal montly installment (EMI) for a loan.  
  
    Arguments:  
        principal - Total amount to be spent (loan + down payment  
        duration - Duration of the loan (in months)  
        rate - Rate of interest (monthly)  
        down_payment (optional) - Optional intial payment (deducted from principal  
    """
    loan_amt = principal - down_payment
    try:
        ans = (loan_amt * rate * (1 + rate)**duration) / ((1 + rate)**duration - 1)
    except ZeroDivisionError:
        ans = loan_amt/duration
    return ans

# no named argument
emi1 = math.ceil(emi(1260000,0.1/12,8*12,300000))
emi2 = math.ceil(emi(1260000,0.08/12,10*12))
# named argument
emi3 = math.ceil(emi(
    principal=1260000,
    rate=.2/12,
    duration=10*12,
    down_payment=200000
))

print("case 1 emi :",emi1,"/month","\ncase 2 emi :",emi2,"/month","\ncase 3 emi :",emi3,"/month")
```

```python
emi_of_home_loan = math.ceil(emi(8e5,.07/12,6*12,200000))
emi_of_car_loan = math.ceil(emi(60000,.12/12,1*12))
total_emi_per_month = emi_of_car_loan + emi_of_home_loan
print("Total emi = ",total_emi_per_month,"/ month")
```

```python
def cost_of_trip(flight_cost,hotel_per_day,day_spend,car_cost_per_week):

    hotel_cost = hotel_per_day * day_spend
    if day_spend >= 7 :
        weeks_spend = int(day_spend / 7)
    else:
        weeks_spend = 1
    car_cost = car_cost_per_week * weeks_spend

    return (hotel_cost + car_cost + flight_cost)
```

```python
# Q1
mumbai = cost_of_trip(
    flight_cost=450,
    hotel_per_day=10,
    day_spend=7,
    car_cost_per_week=70
)

paris = cost_of_trip(
    flight_cost=200,
    hotel_per_day=20,
    day_spend=7,
    car_cost_per_week=200
)

dubai = cost_of_trip(
    flight_cost=370,
    hotel_per_day=15,
    day_spend=7,
    car_cost_per_week=80
)

london = cost_of_trip(
    flight_cost=250,
    hotel_per_day=30,
    day_spend=7,
    car_cost_per_week=120
)

print("Mumbai trip 1 week cost :",mumbai)
print("dubai trip 1 week cost :",dubai)
print("london trip 1 week cost :",london)
print("paris trip 1 week cost :",paris)

```

```python
def phrase_values(data_line):
    values = []
    """ Strip remove the new line character and white spaces. And after that, the split
    will separate each attribute through seperator (,) it is called as item and
    this item in append in values
    """
    for item in data_line.strip().split(','):
        values.append(item)

    return values
```

```python
# create dictionary
# values and headers are list
# both are mapped in dictionary
def create_item_dict(values, headers):
    result = {}
    for value,header in zip(values,headers):
        result[header] = value

    return result
```

### create a csv file and convert to dictionary

```python
def read_csv_file(path):
    result = []
    with open(path,'r') as f:
        # get a list of lines
        lines = f.readlines()
        # parse the header
        headers = phrase_headers(lines[0])
        # loop over remaining lines
        for data_line in lines[1:]:
            # values is list 
            values = phrase_values(data_line)
            # create a dictionary using values & headers
            item_dict = create_item_dict(headers,values)
            result.append(item_dict)
        
    return result
```

```python
read_csv_file('./data/loan2.txt')
```
