### Q print pyramid -> vertical triangle



```python
length = 7
rowSet1 = int(length/2) + 1

for i in range(rowSet1):
    for j in range(i+1):
        print('*',end="")
    print("\n")

rowSet2 = int(length/2) 
for i in range(rowSet2):
    for j in range(rowSet2-i):
        print('*',end="")
    print("\n")
```
